```python
import requests
from bs4 import BeautifulSoup
import time
import pandas as pd
from datetime import datetime
```

```python
req = requests.get('https://habr.com/ru/all/')
```

```python
KEYWORDS = ['python', 'парсинг']
```

```python
soup = BeautifulSoup(req.text, 'html.parser')
posts = soup.find_all('article', class_='post')
```

```python
news = pd.DataFrame()
for post in posts:
    hubs = post.find_all('div', class_='post__text')
    for hub in hubs:
        hub_lower = hub.text.lower()
        for key in KEYWORDS:
            if key in hub_lower:
                heading = post.find('a', class_='post__title_link')
                data = post.find('span', class_='post__time')
                link = heading.attrs.get('href')
                row = {'дата': data.text,'заголовок': heading.text, 'ссылка': link}
                news = pd.concat([news, pd.DataFrame([row])])  
news
```
