```python
log = pd.read_csv('/Users/gost1/Documents/Лекции/доп_материалы_к_пандас_1_и_2/visit_log.csv', sep = ';')
```

```python
def source(data):
    if 'yandex' in data['traffic_source'] or 'google' in data['traffic_source']:
        return 'organic'
    if 'paid' in data['traffic_source'] or 'email' in data['traffic_source']:
        if 'Russia' == data['region']:
            return 'ad'
        else:
            return 'other'
    
    return data['traffic_source']
```

```python
log['source_type'] = log.apply(source, axis=1)
```
