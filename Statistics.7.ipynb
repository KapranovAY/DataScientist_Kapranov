# Задание 1

```python
df = pd.read_csv('/Users/alex/Лекции/Статистика/video_games_sales.csv')
df = df.astype(object).replace(np.nan, 'None')
```

```python
df
```

```python
Sports = df[(df['Genre'] == 'Sports') & (df['Critic_Score'] != 'None')]
```

```python
Other_Genre = df[(df['Genre'] != 'Sports') & (df['Critic_Score'] != 'None')]
```

```python
alpha = 0.05
result = st.ttest_ind(Sports['Critic_Score'], Other_Genre['Critic_Score'], equal_var=False)
print(result)
print(Sports['Critic_Score'].mean())
print(Other_Genre['Critic_Score'].mean())

if (result.pvalue < alpha):
    print('Отвергаем нулевую гипотезу, критики по другому относяться к спортивным играм')
else:
    print('Не отвергаем нулевую гипотезу, критики относяться к жанру Sport, так же как и к остальным жанрам')
```

```python
PC = df[(df['Platform'] == 'PC') & (df['Critic_Score'] != 'None')]
```

```python
PS4 = df[(df['Platform'] == 'PS4') & (df['Critic_Score'] != 'None')]
```

```python
alpha = 0.05
result = st.ttest_ind(PC['Critic_Score'], PS4['Critic_Score'], equal_var=False)
print(result)
print('Средний рейтин PC', PC['Critic_Score'].mean())
print('Средний рейтин PS4', PS4['Critic_Score'].mean())

if (result.pvalue < alpha):
    print('Отвергаем нулевую гипотезу, критики по разному относяться к играм на PC и на PS4')
    if PC['Critic_Score'].mean() > PS4['Critic_Score'].mean():
        print ('Критики больше любят игры на PC')
    else:
        print ('Критики больше любят игры на PS4')
else:
    print('Не отвергаем нулевую гипотезу, критики относяться к жанру Sport, так же как и к остальным жанрам')
```

```python
Shooter = df[(df['Genre'] == 'Shooter') & (df['Critic_Score'] != 'None')]
```

```python
Strategy = df[(df['Genre'] == 'Strategy') & (df['Critic_Score'] != 'None')]
```

```python
alpha = 0.05
result = st.ttest_ind(Strategy['Critic_Score'], Shooter['Critic_Score'], equal_var=False)
print(result)
print('Средний рейтин жанра - Strategy', Strategy['Critic_Score'].mean())
print('Средний рейтин жанры - Shooter', Shooter['Critic_Score'].mean())

if (result.pvalue < alpha):
    print('Отвергаем нулевую гипотезу, критики по разному относяться к жанрам')
    if Strategy['Critic_Score'].mean() > Shooter['Critic_Score'].mean():
        print ('Критики больше Стратегии')
    else:
        print ('Критики больше Шутеры')
else:
    print('Не отвергаем нулевую гипотезу, критики относяться к жанрам одинаково')
```

# Задание 2

```python
df = pd.read_csv('https://raw.githubusercontent.com/obulygin/pyda_homeworks/master/stat_case_study/spam.csv')
```

```python
df['Message'] = df['Message'].str.lower()
```

```python
df['Message'] = df['Message'].str.replace(r'[\W_]+', ' ')
```

```python
from nltk.corpus import stopwords
```

```python
stop = stopwords.words('english')
df['Message'] = df['Message'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
wordnet_lemmatizer = WordNetLemmatizer()
df['Message'] = [' '.join([wordnet_lemmatizer.lemmatize(word) for word in x.split()]) for x in df['Message'].tolist()]
```

```python
df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(df.Message)
names = tfidf.get_feature_names()
tfidf_matrix = pd.DataFrame(tfidf_matrix.toarray(), columns=names)
tfidf_matrix
```

```python
def my_funcion(column): 
    if column['Category'] == 'ham' :
        return 0
    if column['Category'] == 'spam' :
        return 1

df['Tag'] = df.apply(my_funcion, axis=1) 
```

```python
x = tfidf_matrix
```

```python
y = df['Tag'].tolist()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=42)
```

```python
clf = LogisticRegression(random_state=42).fit(x_train, y_train)
```

```python
result = pd.DataFrame([y_test, clf.predict(x_test)]).T
```
