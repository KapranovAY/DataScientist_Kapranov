```python
import numpy as np
from sympy import *
import math
import matplotlib.pyplot as plt
```

```python
from scipy.misc import derivative
```

```python
def f(x):
    return x**5 + 4*sin(2*x) + cos(3*x+3)
derivative(f, 1, dx = 1e-6)
```

```python
x = symbols('x')
gfg_exp = x**5 + 4*sin(2*x) + cos(3*x+3)
dif = diff(gfg_exp, x)
  
print("After Differentiation : {}".format(dif))
```

```python
def f(x):
    return 5*x**4 - 3*sin(3*x + 3) + 8*cos(2*x)
derivative(f, 1, dx = 1e-6)
```

```python
def f(x):
    return x**5 + 4*sin(2*x) + cos(3*x+3)
derivative(f, 1, dx = 1e-6, n=2)
```
