```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
from scipy.stats import norm
from scipy.stats import expon
from scipy.stats import t
```

Уровень 0:

```python
mu = 2.0
sigma = 0.5
norm_rv = sts.norm(loc=mu, scale=sigma)
```

```python
x = np.linspace(0,5,100)
pdf = norm_rv.pdf(x)
cdf = norm_rv.cdf(x)
plt.plot(x, pdf)
plt.plot(x, cdf)

plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.show()
```
