```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
from scipy.stats import norm
from scipy.stats import expon
from scipy.stats import t
```

```python
for  vals  in  [(2.0, 0.5), (3.0, 0.5), (9.0, 0.9), (5.0, 0.2), (6.0, 0.3)]: 
    norm_rv  =  sts.norm(loc=vals[0], scale=vals[1]) 
    expon_rv = sts.expon(loc=vals[0], scale=vals[1])
    t_rv = sts.t(df = 2.74, loc=vals[0], scale=vals[1])
    
    x  =  np.linspace(0, 10, 100) 
    n = norm_rv.pdf(x)
    e = expon_rv.pdf(x)
    t = t_rv.pdf(x)
    
    plt.plot(x, n)
    plt.plot(x, e)
    plt.plot(x, t)
    plt.ylabel('$f(x)$')
    plt.xlabel('$x$')
    plt.show()
```
